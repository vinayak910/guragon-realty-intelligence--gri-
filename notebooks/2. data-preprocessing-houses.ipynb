```python
import re 
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

### Price column

```python
df.price
```

```python
df.price.isnull().sum()
```

```python
df.price.value_counts()
```

```python
df.price.str.split().str[1].value_counts()
```

```python
# 3 types of values -> 1. Crore , Lac , Price on Request 
```

```python
df[df.price == 'Price on Request']
```

```python
# dropping POR rows
```

```python
df = df[df.price != 'Price on Request']
```

```python
def converting_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100 , 2)
        else:
            return round(float(x[0]), 2)
```

```python
df.price = df.price.str.split().apply(converting_price)
```
