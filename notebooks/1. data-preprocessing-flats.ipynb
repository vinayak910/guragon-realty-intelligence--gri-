***function to converting price to numerical (crores)***
- All the values will be representing crores
- ex 56 lac -> 0.56  , 4.5 crore -> 4.5

Here 0.56 is representing 0.56 crore which means 56 lakh , 4.5 is representing 4.5 crores which itself is 4.5 crore

```python
def converting_price(x):
    if type(x) == float :
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100 , 2)
        else :
            return round(float(x[0]), 2)
```

```python
df.price = df.price.str.split().apply(converting_price)
```
